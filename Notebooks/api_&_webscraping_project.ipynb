```python
import requests
import pandas as pd
```

**Getting cryptocurrency prices by market cap via API**

```python
url = "https://api.coingecko.com/api/v3/coins/markets"
```

```python
params = {"vs_currency": "usd","order": "market_cap_desc","per_page": 20,"page": 1,"sparkline": False}
```

```python
response = requests.get(url, params=params)

print("Status Code:", response.status_code)
```

```python
data = response.json()

df_api = pd.DataFrame(data)

df_api.head()
```

**Relevant columns**

```python
df_api = df_api[["name","symbol","current_price","market_cap","total_volume",
"price_change_percentage_24h"]]

```

```python
df_api.columns = ["Name","Symbol","Current Price (USD)","Market Cap","Total Volume","24h Price Change (%)"]

df_api.head()
```

**Creating a csv file of API results**

```python
df_api.to_csv("API DATA", index=False)

```

**Webscrapping list of active crytopcurrencies**



```python
url = "https://en.wikipedia.org/wiki/List_of_cryptocurrencies"

headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)

print(response.status_code)

```

```python

tables = pd.read_html(response.text)

df_web = tables[0]

df_web.head()
```

**Creating a csv file of scraped results from the web**

```python
df_web.to_csv("SCRAPED DATA", index=False)
```
